# Summarytools
### Summarytools is a simple EDA tool that gives a richer summary than `describe()` method. In a single line of code , it generates a standardized and comprehensive data summary .
### The summary includes column statistic,frequency,distribtion chart and missing stats.

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
import pandas as pd
from summarytools import dfSummary

dfSummary(titanic)
```

```python
import pandas as pd
from summarytools import dfSummary

dfSummary(titanic, is_collapsible = True)
```

```python
import pandas as pd
from summarytools import dfSummary, tabset


vaccine = pd.read_csv("C:\\Users\\DESKTOP\\Downloads\\country_vaccinations.csv")
vaccine['date'] = pd.to_datetime(vaccine['date'])

tabset({
    'titanic': dfSummary(titanic).render(),
    'vaccine': dfSummary(vaccine).render()})
```
